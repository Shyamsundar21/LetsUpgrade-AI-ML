```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix,accuracy_score
```

# KNN 

Using the KNN algorithm with different k values to find the one having higher accuracy

```python
df1 = pd.read_csv('Dataset/titanic_train.csv')
df1.head()
```

```python
to_drop = ['PassengerId','Name','Ticket','Cabin']
df1.drop(to_drop,inplace=True,axis=1)
```

```python
le = LabelEncoder()
to_transform = ['Sex','Embarked']
for col in to_transform:
    df1[col] = le.fit_transform(df1[col])
df1.head()
```

```python
y = df1['Pclass']
X = df1.drop('Pclass',axis=1)
```

```python
X_train , X_test , y_train,y_test = train_test_split(X,y,test_size=0.3,random_state = 0)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
def model_fit(k):
    knn = KNeighborsClassifier(n_neighbors=k)
    return knn.fit(X_train,y_train)
```

```python
acc_score_lst = []
for k in range(1,268):
    fitted_model = model_fit(k)
    acc_score = fitted_model.score(X_test,y_test)
    acc_score_lst.append(acc_score)
    y_pred = fitted_model.predict(X_test)
    conf_mat = confusion_matrix(y_test,y_pred)
    print(f"\nk = {k} accuracy_score = {acc_score}")
    print(f"Confusion matrix is\n{conf_mat}")
```

```python
k_val = [i for i in range(1,268)]
plt.plot(k_val,acc_score_lst)
plt.xlabel("K Values")
plt.ylabel("Accuracy score")
plt.title('K value with Accuracy in KNN')
```

## Inference

    -> As the K-value increases the accuracy of the algorithm gets decreased, Hence if k is small the accuracy is high.

    -> We get the accuracy of 89.13% for the k = 1

# SVM

```python
df2 = df1.copy()
```

```python
df2.head()
```

```python
df2.columns
```

```python
from sklearn.svm import SVC
```

```python
def predict_svm(X,y):
    X_train2 , X_test2 , y_train2 , y_test2 = train_test_split(X,y)
    model = SVC(gamma=0.01,C=100)
    model.fit(X_train2,y_train2)
    y_pred = model.predict(X_test2)
    acc = accuracy_score(y_test2,y_pred)
    cmx = confusion_matrix(y_test2,y_pred)
    return acc,cmx
    
```

```python
dv = ['Survived', 'Pclass', 'SibSp', 'Parch','Embarked']
for col in dv:
    y = df2[col]
    X = df2.drop(col,axis=1)
    acc , cmx = predict_svm(X,y)
    print(f"\nDV is {col} Accuracy score = {acc}")
    print(f"Confusion matrix is \n{cmx}")
```

## Inference

The Dependent variables and the accuracy are
        
        1. Survived = 74.88%
        2. Pclass = 91.92%
        3. SibSp = 73.99%
        4. ParCh = 80.71%
        5. Embarked = 73.99%
        
 SVM algorithm has higher accuracy in predicting the Pclass.
